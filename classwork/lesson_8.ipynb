Напишите функцию, которая получает два параметра, <df> и <список столбцов>

Функция должна вернуть сколько данных пропущенно в столбцах из списка, в процентном соотношение.

Input:
func(df, ['Major', 'GPA'])

Output:
Major - 75%
GPF - 30%

```python
def func(df, lst_col):
    sh = df.shape[0]
    for i in lst_col:
        print(f'{i} - {100 - (df[i].dropna().shape[0]*100 // sh)}%')
```

```python
func(df, df.columns)
```
