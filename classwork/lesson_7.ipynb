```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from random import randint, randrange, choice, seed
```

```python
pd.options.display.float_format = '{:.02f}'.format
```

```python
seed(42)

df = pd.DataFrame({
    'student': [f'student{i}' for i in range(10)],
    'subject': [choice(['math', 'english', 'proramming']) for i in range(10)],
    'max_score': [randrange(80, 200, 10) for _ in range(10)],
    'score': [randint(50, 100) for _ in range(10)]
})

df
```

```python
df['persent'] = df[['score', 'max_score']].apply(lambda x:
                                                 min((x['score'] * 100) / x['max_score'], 100),
                                                 axis=1)
```

```python
df_tips = pd.read_csv('tips.csv')
```

```python
df_tips
```

```python
def pan_to_str(pan):
    return str(pan)[:4]
```

```python
df_tips['CC_str'] = np.vectorize(pan_to_str)(df_tips['CC Number'])
```

```python
def is_visa(cart: str) -> str:
    if len(set(cart)) < 4:
        return 'Visa'
    return 'KM'
```

```python
df_tips['cart'] = df_tips['CC_str'].apply(is_visa)
```
