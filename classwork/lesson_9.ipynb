```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.options.display.float_format = '{:.02f}'.format
```

```python
df = sns.load_dataset('mpg')
```

```python
df['mpg'] = df['mpg'] * 1.6 / 3.8, 2
df['displacement'] = df['displacement'] * 16.4 / 5000 * 1000
df['weight'] = df['weight'] * 0.45
df.rename(columns={'mpg': 'kml', 'model_year': 'year'}, inplace=True)
```

```python
df_year = round(df[df['origin'].eq('usa')].groupby('year').mean(numeric_only=True), 2)
```

```python
df_year.loc[[70, 82]]
```

```python
df_name = df.groupby(['name', 'year'], as_index=False).mean(numeric_only=True)
```

```python
df_name = df[df.duplicated(subset='name')].groupby(['name', 'year']).count()
```
