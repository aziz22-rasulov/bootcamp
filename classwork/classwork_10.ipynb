```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
```

```python
plt.figure()
plt.title('Гистограмма')
plt.xlabel('Разгон до (60 миль/час)', fontsize=12)
plt.ylabel('Частота', fontsize=10)
plt.grid(linestyle='--', alpha = 0.9)

plt.axvline(df['acceleration'].mean(), color='red', label='MA')
plt.axvline(df['acceleration'].median(), color='yellow', label='Median')

sns.histplot(df, x='acceleration', bins=20,
            #kde=True,
            #color='purple',
            #element='step',
            #alpha=0.3
            #element='poly',
            #hue='cylinders',
            #palette='Set1'
)
#plt.plot(x, y, marker='o')
plt.legend()
plt.show()
```
