```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('characters_table.csv')
```

```python
df
```

## Fighter

```python
df['Strength'] = df['Strength'].fillna((df['Level'] * 3).where(df['Class'] == 'fighter'))
```

```python
df['Dexterity'] = df['Dexterity'].fillna((df['Level'] * 2).where(df['Class'] == 'fighter'))
```

```python
df['Stamina'] = df['Stamina'].fillna((df['Level'] * 4).where(df['Class'] == 'fighter'))
```

```python
df['Intelligence'] = df['Intelligence'].fillna((df['Level'] * 1).where(df['Class'] == 'fighter'))
```

```python
df['Wisdom'] = df['Wisdom'].fillna((df['Level'] * 3).where(df['Class'] == 'fighter'))
```

```python
df['HP'] = df['HP'].fillna((df['Stamina'] * 15).where(df['Class'] == 'fighter'))
```

```python
df['MP'] = df['MP'].fillna((df['Wisdom'] * 5).where(df['Class'] == 'fighter'))
```

```python
df['DMG'] = df['DMG'].fillna((df['Strength'] * 2).where(df['Class'] == 'fighter'))
```

```python
df['DEF'] = df['DEF'].fillna((df['Level'] * 2).where(df['Class'] == 'fighter'))
```

```python
df['Move'] = df['Move'].fillna((df['Dexterity'] * 1).where(df['Class'] == 'fighter'))
```
